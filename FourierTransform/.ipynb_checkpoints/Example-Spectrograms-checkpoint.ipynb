```python
%matplotlib qt
import numpy as np
import matplotlib.pyplot as plt
import sounddevice as sd
plt.rcParams['figure.figsize'] = [12, 8]
plt.rcParams['font.size'] = 18
```

# Chirp signal

```python
dt = 0.001
t = np.arange(0, 2, dt)
f0 = 50
f1 = 250
t1 = 2
x = np.cos(2 * np.pi * t * (f0 + (f1 - f0) * np.power(t, 2) / (3 * t1 ** 2)))

# fs = 1 / dt
# sd.play(2 * x, fs)
plt.figure()
plt.plot(t, x)
plt.show()
```

```python
plt.specgram?
```

```python
plt.figure()
plt.specgram(x, NFFT=128, Fs=1/dt, noverlap=120, cmap='jet')
plt.colorbar()
plt.show()
```

```python
## Plot the FFT power spectrum
n = len(t)
xhat = np.fft.fft(x, n) # Compute the FFT
PSD = np.real(xhat * np.conj(xhat) / n) # Power spectrum
freq = 1 / (dt * n) * np.arange(n) # Frequencies in Hz (x-axis)
L = np.arange(0, np.floor(n / 2), dtype='int') # Only plot the first half of frequencies

plt.figure()
plt.plot(freq[L], PSD[L], lw=2.5)
plt.xlabel('Frequency, Hz')
plt.show()
```

# Beethoven music

```python
import scipy
import scipy.io
import IPython.display
import librosa
```

```python
# Example to load mat file (first 40 seconds)
file = scipy.io.loadmat('/Users/creator/data/databookuw/beethoven_40sec.mat')
sound = file["y"][0] # data
T = int(file["T"][0][0]) # length in seconds
# FS = int(file["T"][0][0]) # sample frequency
FS = 24000 # sample frequency
samples = T * FS
t = np.arange(samples)

plt.figure(figsize=(20, 5))
plt.plot(t, sound)
plt.show()

IPython.display.Audio(sound, rate=FS)

plt.figure(figsize=(20, 5))
plt.specgram(sound, NFFT=5000, noverlap=400, Fs=FS, cmap='jet')
plt.colorbar()
plt.show()
```

```python
# original mp3 file
FS = 24000
y, sr = librosa.load('/Users/creator/data/databookuw/beethoven.mp3')

plt.figure(figsize=(20, 5))
plt.plot(y)

IPython.display.Audio(y, rate=FS)

plt.rcParams['figure.figsize'] = (20, 5)
plt.figure()
plt.specgram(y, NFFT=8192, Fs=FS, noverlap=2048)
plt.colorbar()
plt.show()
```
